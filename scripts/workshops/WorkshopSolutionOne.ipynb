```python
import numpy as np
import matplotlib.pyplot as plt

w = np.array([3, -4])
b = 2

x2 = np.array([0, 1])
x1 = -(w[1]*x2/w[0])-(b/w[0])
print(x1)

fig, ax = plt.subplots()

ax.plot(x2, x1, linewidth=2.0)
ax.scatter(x=np.array([0.2, 0.6]), y=np.array([0.4, -0.4]), c=['yellow','red'])
plt.show()
```
